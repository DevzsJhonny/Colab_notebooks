```python
# e por fim, vamos finalizar esse grafico da melhor forma
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


axis = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por Unidade da Federação")
plt.show()
```

```python
# DESAFIO 3 - reposicionar legenda atraves do T
# colocar titulos nos dois eixos
```

```python
import matplotlib.pyplot as plt
import numpy as np

# para criar o grafico com base no T, reposicionando as linhas e colunas
dados_usaveis.T.plot(figsize=(15,9))

# para reposicionar a legenda
#plt.legend(loc=4)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left',borderaxespad=0)

# para criar titulos aos eixos X e Y
plt.xlabel("Meses Atendimento",size=14)
plt.ylabel("Valor", size=14, rotation=40)

plt.title("Valor por unidade de Federação")
plt.show()
```
