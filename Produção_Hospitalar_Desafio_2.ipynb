```python
import pandas as pd
```

```python
dados = pd.read_csv('A150850189_28_143_208 (1).csv', encoding='ISO-8859-1', skiprows=3,skipfooter=12, sep=';',  thousands='.', decimal=',')

## desafio 1 fiap, mostrar a coluna mais recente do arquivo
#dados = pd.read_csv('/content/A152918189_28_143_208.csv', encoding='ISO-8859-1', skiprows=3, sep=';',  thousands='.', decimal=',')
```

```python
pd.options.display.float_format = "{:.2f}".format #notação python
```

```python
dados
```

```python
#primeira tentiva de usar o plot para criação de grafico (matplotlib ja intalado no Colab)
dados.plot(x='Unidade da Federação', y='2008/Ago', figsize=(9,6))
```

```python
# dessa vez vamos fazer um grafico do tipo de Barras, mudando o tamanho que é apresentado
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
```

```python
# e por fim, vamos finalizar esse grafico da melhor forma
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


axis = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por Unidade da Federação")
plt.show()
```
