```python
def plota_gastos_por_mes(dados, figsize=(10,6)):

  axis = dados.T.plot(figsize=figsize)
  axis.set_ylabel("gastos aprovado em milhões de reais", size=10)
  axis.set_xlabel("Meses")
```
