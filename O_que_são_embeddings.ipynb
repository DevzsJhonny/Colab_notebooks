```python
import gensim

from gensim.models import Word2Vec
```

Embeddings nada mais são do que as representações numéricas de palavras!

* **IMAGINE QUE O ÍCONE VERMELHO É UMA ENTRADA NOVA**
* **CADA VETOR (CIRCULO BRANCO, LARANJA, E AZUL) É UMA REPRESENTAÇÃO NUMÉRICA DE UMA DETERMINADA STRING**

image.png

```python
# pode ser um array com itens de N caracteres também
sentences = [
    ["Eu", "quero", "café", "pequeno"],
    ["Eu", "quero", "café", "extra", "grande"],
    ["Eu", "quero", "café", "médio"]
]
```

```python
model = Word2Vec(sentences, vector_size=3, window=3, min_count=1, sg=1)
model.save("word2vec.model")
model = Word2Vec.load("word2vec.model")
```

Olha só como que fica, quando transformamos string para embedding usando o Word2Vec:

```python
model.wv.__dict__
```

```python
print("Embedding para 'café' (representeção numérica da palavra 'café'):", model.wv['café'])
```

```python
print("Palavras próxmias de 'café':", model.wv.most_similar('café', topn=3)) # topn = 3 É TIPO O KNN, ENTÃO ELE VAI PEGAR OS 3 MAIS PRÓXIMOS DA PALAVRA DE ENTRADA
```
