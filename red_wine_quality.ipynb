```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dados = pd.read_csv('/content/winequality-red.csv', encoding='ISO-8859-1', sep=';',   decimal=',')
pd.options.display.float_format = "{:.2f}".format
dados
```

```python
# verificando linhas duplicadas
dados.duplicated().sum()
```

```python
# Removendo linhas duplicadas
dados = dados.drop_duplicates()

# Verificando se existem valores nulos
dados.isnull().sum()

```

```python
# verificando linhas duplicadas novamente
dados.duplicated().sum()
```

```python
#renomear colunas
dados = dados.rename(columns={'fixed acidity': 'acidez','volatile acidity': 'acidez volatil','citric acid':'acidez citrica', 'residual sugar': 'acucar contido','chlorides':'cloreto', 'free sulfur dioxide': 'dioxido sulfurico', 'total sulfur dioxide':'dioxido sulfurico total', 'density': 'densidade','pH':'pH','sulphates':'sulfatos','alcohol':'teor alcolico','quality':'qualidade'})
dados
```

```python
dados.info()
```

```python
dados.dtypes
```

# Usando matplotlib para:


**distribuição da acidez volátil por qualidade de vinho**

```python


grafico_px = px.area(dados, x='acidez volatil', y='qualidade', title='acidez volátil por qualidade de vinho')
grafico_px.show()
```

**Distribuição do teor alcóolico por qualidade do vinho**

```python
grafico_px_new = px.scatter(dados, x='teor alcolico', y='qualidade', title='teor alcóolico por qualidade do vinho')

grafico_px_new.show()
```

**Relação entre acidez volátil e teor alcóolico**

```python
grafico_px_teor = px.scatter(dados.sort_values(by='acidez volatil', ascending=False), x='acidez volatil', y='teor alcolico', title='acidez volatil e teor alcóolico')

grafico_px_teor.show()
```

**Relação entre teor alcóolico e qualidade do vinho**

```python
grafico_px_qua = px.histogram(dados.sort_values(by='qualidade', ascending=False), x='teor alcolico', y='qualidade', title='teor alcóolico x qualidade do vinho')
grafico_px_qua.show()
```

```python
sns.jointplot(x='teor alcolico', y='qualidade', data=dados)
```
